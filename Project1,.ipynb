```python
import pandas as pd 
billboard = pd.read_csv('billboard.csv')
billboard
```

```python
week_list = list(billboard.columns)[5:]
tag_list = list(billboard.columns)[0:5]
```

```python
billboard_new = pd.melt(billboard, id_vars = tag_list , value_vars = week_list, 
        var_name = 'week', value_name = 'rank')
```

```python
billboard_new['week'] = billboard_new['week'].str.extract('(\d+)').astype(int)
billboard_new
```

```python
billboard_new.dtypes
```

```python
billboard_new['date.entered'] = pd.to_datetime(billboard_new['date.entered'])
billboard_new['date.peaked'] = pd.to_datetime(billboard_new['date.peaked'])
```

```python
billboard_new.dtypes
```

```python
billboard_new
```

```python
billboard_new = billboard_new.assign(entered_to_peak = billboard_new['date.peaked']-billboard_new['date.entered'])
```

```python
billboard_new['entered_to_peak'] =  billboard_new['entered_to_peak'].dt.days
```

```python
billboard_new.dtypes
```
